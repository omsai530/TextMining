```python
import nltk
import pandas as pd
import numpy as np
import re
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
Lemmer = WordNetLemmatizer()
Stemmer = PorterStemmer()
```

```python
def My_Stemmer(Str):
    words=re.sub(r'[^A-Za-z\-]','',Str).lower().split()
    words=[Stemmer.stem(word) for word in words]
    return words
```

```python
df = pd.read_csv("Countv.csv")
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df = df.sample(2000, random_state=42)
```

```python
df
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
```

```python
Model_NB1= MultinomialNB()
```

```python
TrainDF, TestDF = train_test_split(df, test_size=0.3)
```

```python
TrainDF
```

```python
TestDF
```

```python
Train_labels = TrainDF['LABEL']
```

```python
Test_labels  = TestDF['LABEL']
```

```python
TrainDF.drop(columns='LABEL', inplace=True)
TestDF.drop(columns='LABEL', inplace=True)
```

```python
Model_NB1.fit(TrainDF, Train_labels)
```

```python
Pred=Model_NB1.predict(TestDF)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cnf_mat = confusion_matrix(Test_labels, Pred)
```

```python
class_names=df['LABEL'].unique().astype(str)
heatmap=sns.heatmap(cnf_mat, annot=True, cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
ac1 = accuracy_score(Test_labels, Pred)
```
