```python
import preprocessor as p
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Final (1).csv')
```

```python
df.head(5)
```

```python
new_tweets=[]
for i in df['tweet']:
  new_tweets.append(p.clean(str(i)))
```

```python
import nltk
from nltk import word_tokenize, FreqDist
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download
nltk.download('wordnet')
nltk.download('stopwords')
from nltk.tokenize import TweetTokenizer
```

```python
import re
for i in range(0,len(new_tweets)):
  new_tweets[i]= new_tweets[i].replace('\d+', '')
  new_tweets[i] = new_tweets[i].lower()
  new_tweets[i]=re.sub(r'[\"\,\.\?\:]', '', (new_tweets[i]))
  print(i)
```

```python
import nltk
nltk.download('omw-1.4')
stop_words = set(stopwords.words('english'))
w_tokenizer = TweetTokenizer()
for i in range(0,len(new_tweets)):
  w=w_tokenizer.tokenize(str(new_tweets[i]))
  new_tweets[i]=w
  new_tweets[i] = [item for item in new_tweets[i] if item not in stop_words]
```

```python
from nltk.stem.porter import PorterStemmer
l=new_tweets
```

```python
import gensim
from gensim import corpora
dictionary = corpora.Dictionary(l)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in l]
```

```python
Lda = gensim.models.ldamodel.LdaModel
ldamodel = Lda(doc_term_matrix, num_topics=7, id2word = dictionary, passes=10)
```

```python
#Topic Table
print(ldamodel.print_topics(num_topics=7, num_words=10))
```

```python
import pickle
with open("model.pkl","wb") as handle:
  pickle.dump(ldamodel,handle)
```
