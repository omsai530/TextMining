```python
import pandas as pd
import nltk
nltk.download('omw-1.4')
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import TweetTokenizer
```

```python
df = pd.read_csv('Final.csv')
df
```

```python
df['tweet']
```

```python
df_arm = pd.DataFrame(df['tweet'])
df_arm
```

```python
new_tweets = df['tweet']
```

```python
stop_words = set(stopwords.words('english'))
w_tokenizer = TweetTokenizer()
for i in range(0,len(new_tweets)):
  w = w_tokenizer.tokenize(str(new_tweets[i]))
  new_tweets[i] = w
  new_tweets[i] = [item for item in new_tweets[i] if item not in stop_words]
  new_tweets[i] = [item for item in new_tweets[i] if len(item)>2 ]
```

```python
transactions_list = new_tweets.tolist()
transactions_list
```

```python
import csv

# Save transaction list as CSV file
with open('Final_ARM.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(transactions_list)
```
